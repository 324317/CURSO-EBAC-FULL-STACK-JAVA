```python
import pandas as pd
import numpy as np
import math 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Use_Of_Force.csv')
df
```

```python
df.columns
```

```python
df ['Occured_date_time'] = pd.to_datetime(df['Occured_date_time'],format='%m/%d/%Y %I:%M:%S %p')
df
```

```python
df2 = df.sort_values('Occured_date_time',ascending = True,ignore_index=True)
df2
```

```python
# OCORRE ERRO (CORRIGIR)
df2['Occured_date_time'] = df2['Occured_date_time'].dt.round(freq='Y')
```

```python
df2.drop_duplicates(subset=('Occured_date_time','Subject_ID'), keep='first', ignore_index=True,inplace=True)
df2
```

```python
agr_year = df2.groupby(df2['Occured_date_time'].dt.year, sort=False)['Incident_Num'].count()
agr_year
```

```python
plt.title('Crimes por Ano', size = 20)
agr_year.plot.bar()
```

```python
agr_month = df2.groupby(df2['Occured_date_time'].dt.month, sort=False)['Incident_Num'].count()
agr_month
```
